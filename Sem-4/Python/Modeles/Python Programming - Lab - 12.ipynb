### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
x = ["A", "B", "C", "D"]  
y = [1, 2, 3, 4]          
plt.bar(x, y)
for index, value in enumerate(y):
    plt.text(x[index], value + 0.08, str(value),fontsize=10)

plt.show()
```
